````python
# DEFINICION DE LA CLASE

class TiendaOnline:
    def __init__ (self, ventas_totales):
        
        self.inventario = [] # List of dictionaries of products in the inventory: {producto: 'nombre', 'precio', 'cantidad'}
        self.clientes = {} # dictionary to register clients: 
        self.ventas_totales = ventas_totales # to carry total sales   
    

    # 1. añadir método para agregar productos o actualizar cantidades
    # producto = {'nombre':"nombre", 'precio':precio, 'cantidad' = cantidad}
    def agregar_producto(self, nombre, precio, cantidad):
        # iterar a través del inventario. 
        # Para cada diccionario_producto en self.inventario 
        for producto in self.inventario: 
            # comprobar si el nombre de producto proporcionado se corresponde con algún producto en el inventario 
            if producto['nombre'] == nombre: 
                # si hay concordancia se actualiza la cantidad
                producto['cantidad'] += cantidad
                print(f'Se ha actualizado la cantidad del producto {nombre}') 
                #se devuelve el inventario actualizado                
                return self.inventario    
        # si el producto es nuevo se añade al diccionario con sus pares llave:valor  
        producto_nuevo = {'nombre':nombre, 'precio':precio, 'cantidad':cantidad}
        self.inventario.append(producto_nuevo)
        print(f'Se ha añadido el producto {nombre}')
        

    # 2. añadir un método que muestre el inventario de productos con sus detalles
    def ver_inventario(self): 
        print("\nInventario de productos:")
        print("----------------")
    # iterar en todos los productos del inventario
        for producto in self.inventario:
        # imprimir, para cada producto, los contenidos de las llaves Nombre, Precio, Cantidad
            print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")
    

    # 3. añadir un método para buscar un producto en el inventario por nombre y muestra sus detalles si se encuentra
    def buscar_producto(self, nombre):
        # usamos un bucle for para recorrer el inventario
        for producto in self.inventario:
            # si se encuentra el producto:
            if producto['nombre'] == nombre:            
                # imprime sus detalles
                print('Los detalles del producto que estas buscando son:')
                print(f"Nombre: {producto['nombre']}, Precio: {producto['precio']}, Cantidad: {producto['cantidad']}")    


    # 4. actualizar stock, sumar o restar la cantidad que entra como argumento
    def actualizar_stock(self, nombre, cantidad):
        cantidad_actualizar = cantidad
        # usamos un bucle for para recorrer el inventario
        for producto in self.inventario:
            if producto['nombre'] == nombre:               
                producto['cantidad'] += cantidad_actualizar
                print(f"\nEl stock del producto {producto['nombre']} se ha actualizado. El stock ahora es {producto['cantidad']}")
                break
        else: 
            print(f'El producto {nombre} no está en el inventario')                    

    
    # 5. metodo para eliminar productos
    def eliminar_producto(self, nombre):
        # usamos un bucle for para recorrer el inventario
        for producto in self.inventario:
            # comprobar si el producto está en el inventario
            if producto['nombre'] == nombre: 
                # en caso afirmativo eliminar todo el diccionario producto del inventario
                self.inventario.remove(producto)
                print(f"El producto {nombre} se ha eliminado")
                return self.inventario # devolver el inventario modificado
            # en caso de que el producto no esté en el inventario
        else: 
            print(f'El producto {nombre} no está en el inventario')
            return self.inventario # devolver el inventario sin modificaciones
              

    # 6. metodo para calcular el valor total del inventario
    def calcular_valor_inventario(self):
        valor_total_inventario = 0
        # usamos un bucle for para recorrer el inventario
        for producto in self.inventario:
            valor_producto = producto["precio"] * producto["cantidad"]
            valor_total_inventario += valor_producto
        print(f'\nEl valor total del inventario es {valor_total_inventario}')

    

    # 7. metodo para realizar una compra
    # En este método se llama a los métodos anteriores ver_inventario(self) y actualizar_stock(self, nombre, cantidad)
    #[comentario para mi] Versión más sencilla de la opcion_2 

    def realizar_compra(self):
        compra_realizada = False
        carrito_compra = []
        coste_compra = 0

        while not compra_realizada:
            pregunta_compra = input("¿Quiere realizar una compra? si o no")
            if pregunta_compra.lower() == "si":                
            # se llama a la función ver_inventario para mostrar el inventario al cliente
                self.ver_inventario()

                # preguntar al cliente qué producto quiere comprar y cuantas unidades
                producto_compra = input("¿Qué producto quieres comprar?:")                        
                unidades_producto_compra = int(input("¿Cuantas unidades de ese producto quieres comprar?:"))
                # como esta dos preguntas están fuera del bucle que comprueba si el producto ya está en el carrito, 
                # pregunta igual cuantas unidades quieres antes de informar que el producto ya está en el carrito

                for producto in self.inventario:                   

                    if (producto_compra == producto['nombre'].lower()) and (producto['cantidad'] >= unidades_producto_compra):
                        # comprobar si el producto ya está en el carrito
                        if producto_compra in carrito_compra:
                            print("Ese producto ya está en el carrito")
                            # pregunta si se quieren modificar las unidades
                            pregunta_añadir_unidades = input("¿Quieres modificar el número de unidades?")
                            if pregunta_añadir_unidades.lower() == "si": 
                                añadir_o_restar = input("¿Quieres añadir o restar unidades?: ")
                                if añadir_o_restar.lower() == 'añadir':
                                    unidades_añadir = int(input('¿Cuantas unidades?'))
                                    unidades_producto_compra += unidades_añadir
                                    coste_compra += (unidades_añadir*precio_producto)
                                    # aquí también hay que actualizar el stock
                                    # llamando al método actualizar_stock                                    
                                    self.actualizar_stock(producto_compra, -unidades_añadir) 
                                    
                                elif añadir_o_restar.lower() == 'restar':
                                    unidades_restar = int(input('¿Cuantas unidades?'))
                                    unidades_producto_compra -= unidades_restar
                                    coste_compra -= (unidades_restar*precio_producto)
                                    # aquí también hay que actualizar el stock
                                    # llamando al método actualizar_stock
                                    self.actualizar_stock(producto_compra, +unidades_restar)
                                    # habría que añadir otro condicional de manera que si finalmente las unidades de ese producto
                                    # en el carrito son 0, lo elimine del carrito. 
                                                                    
                        
                        else:
                        #se añade el producto al carrito
                            carrito_compra.append(producto_compra)
                            precio_producto = producto['precio']
                            # el coste de la compra es el precio del producto por el número de unidades
                            coste_compra += (precio_producto*unidades_producto_compra)
                            # aquí se actualiza el stock, llamando al método actualizar_stock
                            #stock_actualizado = producto['cantidad'] - unidades_producto_compra
                            self.actualizar_stock(producto_compra, -unidades_producto_compra) 
                        print(f'\nTu carrito de la compra contiene: {set(carrito_compra)} y el coste total es {round(coste_compra, 2)}')                      
                        break # salir del bucle si se encuentra el producto 
                    
                else: #(producto_compra != producto['nombre'].lower()) or (producto['cantidad'] < unidades_producto_compra):
                    print("Ese producto no está disponible en este momento")           


            else:
                # finalizar el bucle while
                compra_realizada = True
                print("\nVamos a procesar el pago de su compra")
                print("----------------------")



    # 8. método para procesar pago
    def procesar_pago(self):
        # pedir al cliente la cantidad total que debe pagar
        cantidad_total = float(input("¿Cual es la cantidad que debe pagar?: "))
        cantidad_pagada = float(input("¿Qué cantidad ha introducido para pagar?: "))

        print(f"\nLa cantidad que debe pagar es {round(cantidad_total, 2)}€ y ha introducido {round(cantidad_pagada, 2)}€")

        # comprobar que la cantidad pagada es superior a la total
        if cantidad_pagada > cantidad_total:
            cambio = cantidad_pagada - cantidad_total
            print(f'\nAquí tiene su cambio: {round(cambio,2)}€, que tenga un buen día, esperamos volver a verle pronto')
        elif cantidad_pagada == cantidad_total:
            #cambio = cantidad_pagada - cantidad_total
            print(f'\nMuchas gracias, que tenga un buen día, esperamos volver a verle pronto')
        elif cantidad_pagada < cantidad_total: # si no es el caso, pedir que se pague la diferencia 
            falta = cantidad_total - cantidad_pagada
            print(f'\nLo que ha pagado no es suficiente, debe añadir {falta}€ para pagar la totalidad de su compra')


    # 9. método agregar cliente
    def agregar_cliente(self, nombre, email):        
    # verificar si ese cliente ya existe en self.clientes
        if nombre not in self.clientes.keys():
            # agregar cliente al diccionario
            self.clientes[nombre] = {
                'email': email,
                'compras': []
            }   
            print(f'El cliente {nombre} se ha agregado con éxito')
            #return self.clientes
        else: #nombre in self.clientes:
            print(f'El cliente {nombre} ya se encuentra en la base de datos')      
                   
   # 10. método ver clientes
    def ver_clientes(self):
        for key in self.clientes.keys():
            # para conocer el nombre del cliente, que es la 1ª llave del dict self.cliente
            # la funcion iter() itera sobre todas las llaves del diccionario
            # la función next() se queda con la primera llave
            cliente_nombre = key 
            cliente_info = self.clientes[cliente_nombre]

            #print(f'Nombre: {cliente_nombre}, email: {cliente_info['email']}')
            print('Nombre: ', cliente_nombre, "; email:", cliente_info['email'])

    
    # He conseguido realizar hasta aquí, los otros métodos no me ha dado tiempo a intentarlos       





````
